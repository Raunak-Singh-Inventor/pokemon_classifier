Original dataset from: https://www.kaggle.com/datasets/lantian773030/pokemonclassification

Divide into train and test sets

```python
import zipfile

def unzip_data(data_dir):
  z = zipfile.ZipFile(data_dir)
  z.extractall()
```

```python
import os

data_dir = "PokemonData/"
for label in os.listdir(data_dir):
    filenames = os.listdir(data_dir + label)
    train_len = .8 * len(filenames)
    !mkdir pokemon_data_modified/train/{label}/
    !mkdir pokemon_data_modified/test/{label}/
    for i, image in enumerate(filenames):
        new_path = data_dir + label + "/" + str(i) + ".jpg"
        print("Image to move: " + str(new_path))
        !mv {data_dir + label + "/" + image} {new_path}
        image_path = new_path
        if i<=train_len:
            !cp {image_path} pokemon_data_modified/train/{label}/
        else:
            !cp {image_path} pokemon_data_modified/test/{label}/
```

```python
mod_data_dir = "pokemon_data_modified/"
for (dirpath, dirnames, filenames) in os.walk(mod_data_dir):
    print(f"There are {len(dirnames)} directories and {len(filenames)} images in '{dirpath}'.")
```
